# 3. Visual Analysis

```python
#univariate analysis for number of product purchased
plt.figure(figsize=(11, 4.5))
sns.countplot(data=df,x="Product")
```

**Insights**: Countplot shows that KP281 is purchased most by customer and customer seems to like the economical ones.

**Recommendation**: In an emerging markets like india, people prefer value for money, if more features are introduced to KP281, then the sales will go high and aerofit should foucs on this.

```python
# Marital status vs number of produt purchased
plt.figure(figsize=(11, 4.5))
sns.countplot(data=df,x="MaritalStatus",hue="Product")
```

**Insights**: Countplot shows that partnered people buy the products more than single.  Using both salaries to buy the products may also the reason for this hike.  KP281 tops in both categories while KP781 bottoms

**Recommendation**: To increase the sales of KP781, aerofit may target people who are partnered and salaried.  They also can provide flexible EMIs to increase sales

```python
#Marginal probability for gender and product
round(pd.crosstab(index=[df.Gender,df.MaritalStatus],columns=df['Product'],margins=True,dropna=True,normalize='all')*100,2)
```

**Insights**:

  Marginal probability shows than KP281 **purchase double** when customer is **female and partnered**.  

  Also it **doubles** in KP481 where the **male is partnered**.  

  A 35 % increase in sales of KP781 between single and partnered people.

  This shows that woman though tend to buy product considers the price whereas male goes 1 step above.

**Recommendation**:

  Sales team must focus on partnered females for KP281 and Partnered males for KP481

```python
#Age vs number of product purchased
plt.figure(figsize=(11, 4.5))
sns.histplot(data=df,x="Age",hue="Product",binwidth=1,multiple="stack")
```

**Insights:**  The histplot shows that KP281 is bought by almost all age groups.
KP781 is abundantly used between age group 23 to 32
All the products are heavy sold in age: 26 category.
The graph starts at age 18 and slowly subside on age 36

**Recommendation**: Sales should target age group 20 to 30 to increase profit margins

```python
#boxplot analysis for finding outliers
#bivariate analysis, #Outliers
# age vs product
plt.figure(figsize=(11, 4.5))
sns.boxplot(data=df,x="Age",y="Product")
```

**Insights**: The box plot shows that KP281 is the most liked product among all age groups followed by KP481 and KP781.  The median age also falls between 25 to 30 years. KP781 has got more outliers among 3 products.  The chart length between 75% to 100% is greater for all products.  Upper quartile is high for all products.

**Recommendation**:  Aerofit should focus on less outlier product

```python
#income vs product
plt.figure(figsize=(11, 4.5))
sns.boxplot(data=df,x="Income",y="Product")
```

**Insights**: Boxplot shows that people with higher salary tend to buy KP781 more whereas in KP281 its viceversa.  KP281 and KP481 are preferred by mid range salaried people

**Reccomendation**:  We can aim to sell other aerofit premium products to high salaried people by giving specicalized offers.

Whereas give EMI support to mid range salaried people to sell KP481 more.

```python
#Education vs usage
plt.figure(figsize=(11, 4.5))
sns.boxplot(data=df,x="Education",y="Usage")
```

**Insights**: Box plot shows that people with 18 years of education are perfect usage people.  This shows that they are health conscious.

```python
#Crosstab
pd.crosstab(index=df['Product'],columns=df['Gender'],margins=True,dropna=True)
```

**Insights**:  KP281 and KP481 are shared by male and female equally whereas KP781 is shared by mostly males.

```python
#Conditional probability
round(pd.crosstab(index=[df.Gender,df.MaritalStatus],columns=df['Product'],margins=True,dropna=True,normalize='index')*100,2)
```

```python
#usage vs product
round(pd.crosstab(index=df['Usage'],columns=df['Product'],margins=True,dropna=True,normalize='all')*100,2)
```

**Insights**: KP281 and KP481 is used mostly used by people who use treadmill thrice a week.

KP781 is mostly used by people who use treadmill four times a week.

KP281 and KP481 aren't used by people who told they would use the treadmill for 6 and 7 times a week.

KP781 isn't used by people who said would use the product twice a week.

```python
#Fitness vs product
round(pd.crosstab(index=df['Fitness'],columns=df['Product'],margins=True,dropna=True,normalize='index')*100,2)
```

**Insights**: KP281 is mostly used by mid or intermediate fitness people.  KP481 is mostly used by low fitness or beginner people and KP781 is used by high fitness or expert people.
