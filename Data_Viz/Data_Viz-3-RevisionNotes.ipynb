Coin toss experiment

```python
#Simulting a coin toss
#Heads: 1
#Tails: 0

np.random.choice([0,1])
```

```python
#Simulating an experiment where I toss 10 coins in a row
#Fair coin

n = 1_000_000 #no of people tossing 10 coins in a row
outcomes = np.random.choice([0,1],size=(n,10),p=[0.5,0.5])

print(outcomes)

no_of_heads = outcomes.sum(axis=1)
print("\n\nNo of heads:",no_of_heads)
```

```python
#plotting the no of heads in above exp

sns.countplot(x=no_of_heads)
```

Coin toss with a biased coin

```python
#Simulating an experiment where I toss 10 coins in a row
#Biased coin: Heads 80% of the times

n = 1_000_000 #no of people tossing 10 coins in a row
outcomes = np.random.choice([0,1],size=(n,10),p=[0.2,0.8])

print(outcomes)

no_of_heads_biased = outcomes.sum(axis=1)
print("\n\nNo of heads:",no_of_heads_biased)
```

```python
sns.countplot(x=no_of_heads_biased)
```

Using the sns "Penguin" dataset

```python
#loading the 'Penguin' dataset from seaborn

df = sns.load_dataset('penguins')

df
```

```python
#uni-variate analysis of body_mass_g

sns.histplot(data=df,x='body_mass_g')
```

```python
#on previous uni-variate analysis of body_mass_g
#add another variable -> 'species'

sns.histplot(data=df,x='body_mass_g',hue='species')
```

```python
#on previous uni-variate analysis of body_mass_g
#add another variable -> 'species'
#kdeplot -> Kernel "Density" Estimate

sns.kdeplot(data=df,x='body_mass_g',hue='species')
```

```python
#on previous uni-variate analysis of body_mass_g
#add another variable -> 'species'
#kdeplot -> Kernel "Density" Estimate
#wider figure

plt.figure(figsize=(12,4))

#set cursor for subplot 1: Males
plt.subplot(1,2,1)
sns.kdeplot(data=df[df['sex']=='Male'],x='body_mass_g',hue='species')
plt.title("Male Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)


#set cursor for subplot 2: Females
plt.subplot(1,2,2)
sns.kdeplot(data=df[df['sex']=='Female'],x='body_mass_g',hue='species')
plt.title("Female Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)


plt.show()
```

```python
#adding a 3rd subplot

plt.figure(figsize=(12,10))

#set cursor for subplot 1: Males
plt.subplot(2,2,1)
sns.kdeplot(data=df[df['sex']=='Male'],x='body_mass_g',hue='species')
plt.title("Male Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)


#set cursor for subplot 2: Females
plt.subplot(2,2,2)
sns.kdeplot(data=df[df['sex']=='Female'],x='body_mass_g',hue='species')
plt.title("Female Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)

#set cursor for subplot 3: Males
plt.subplot(2,2,3)
sns.kdeplot(data=df[df['sex']=='Male'],x='body_mass_g',hue='island')
plt.title("Male Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)

#set cursor for subplot 4: Female
plt.subplot(2,2,4)
sns.kdeplot(data=df[df['sex']=='Female'],x='body_mass_g',hue='island')
plt.title("Female Penguin weights")
plt.xlim(2000,7000)
plt.ylim(0,0.0006)


plt.show()
```
