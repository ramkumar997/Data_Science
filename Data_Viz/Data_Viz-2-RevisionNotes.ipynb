Coin toss experiment

```python
#Simulting a coin toss
#Heads: 1
#Tails: 0

np.random.choice([0,1])
```

```python
#Simulating an experiment where I toss 10 coins in a row
#Fair coin

n = 1_000_000 #no of people tossing 10 coins in a row
outcomes = np.random.choice([0,1],size=(n,10),p=[0.5,0.5])

print(outcomes)

no_of_heads = outcomes.sum(axis=1)
print("\n\nNo of heads:",no_of_heads)
```

```python
#plotting the no of heads in above exp

sns.countplot(x=no_of_heads)
```

Coin toss with a biased coin

```python
#Simulating an experiment where I toss 10 coins in a row
#Biased coin: Heads 80% of the times

n = 1_000_000 #no of people tossing 10 coins in a row
outcomes = np.random.choice([0,1],size=(n,10),p=[0.2,0.8])

print(outcomes)

no_of_heads_biased = outcomes.sum(axis=1)
print("\n\nNo of heads:",no_of_heads_biased)
```

```python
sns.countplot(x=no_of_heads_biased)
```
