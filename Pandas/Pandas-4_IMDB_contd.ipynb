### Group based fiiltering

```python
# WHERE - filters the rows before grouping
# HAVING - filter the group based rows that are created, filter after the group
# Want to filter the rows of the entire group(s) but from the original data, not the grouped data
```

```python
# Find the rows (movies) which are directed by high budget director (>$ 100M)

# Ques1. Filter the high budget movies - budget > $100
# Ques2. Filter the movies which are directed by high budget directors
```

```python
# group based filtrerinmg - filter the rows of original (not grouped data) based on group level properties
data.groupby("director_name").filter(lambda x: x["budget"].max() >= 100000000)
# x --> dataframe corresponding to one of the groups
# x["budget"].max() calculating the budget of most expensive movie of a pariticular group/director
```

### Group based apply

Risky Movies (for a director) - Expected Budget of this movies is > average revenue from the director's other movies

```python
def func(x):
  x["risky"] = x["budget"] - x["revenue"].mean() >= 0
  return x

data_risky = data.groupby("director_name").apply(func)
data_risky
# transform - similar to apply functiomn
```

```python
# Number of movies/ Number of Active Year (max_year- min_year)
data_agg = data.groupby(["director_name"])[["title", "year"]].aggregate({"title": "count",
                                                              "year":["min", "max"]})
data_agg
```

```python
data_agg.columns
```

```python
data_agg["title"]
```

```python
data_agg["title"]["count"]
```

```python
data_agg["year"]["min"]
```

```python
data_agg.columns = ['_'.join(col) for col in data_agg.columns]
data_agg
```

```python
data_agg.reset_index()
```

```python
data_agg["yrs_active"] = data_agg["year_max"] - data_agg["year_min"]
```

```python
data_agg
```

```python
data_agg["movie_per_yr"] = data_agg["title_count"] / data_agg["yrs_active"]
data_agg
```

```python
data_agg.sort_values("movie_per_yr", ascending=False)
```
